```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
from langchain import HuggingFacePipeline

model_id = "meta-llama/Llama-3.2-3B-Instruct"

base_model = AutoModelForCausalLM.from_pretrained(model_id,
                                                  load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained(model_id)

pipe = pipeline(
    "text-generation",
    model=base_model,
    tokenizer=tokenizer,
    max_new_tokens=250,
    temperature=0.7,
    top_p=0.95,
    repetition_penalty=1.15
    )
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
ask = f"""Below is an instruction that describes a task. Write a response that appropriately completes the request.
### Instruction:
suggest two ways two lose weight.

Answer:"""
```

```python
text = llm(ask)
```

```python
print(text.split("Answer:")[1].strip())
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory

prompt = PromptTemplate(
    input_variables=["history", "input"],
    template="""
You are an assistant. Continue the conversation below:

{history}
Human: {input}
AI:"""
)

memory = ConversationBufferMemory()

conversation = LLMChain(
    llm=llm,
    prompt=prompt,
    memory=memory,
    verbose=False
)
```

```python
output_1 = conversation.predict(input="Can you list 3 places in USA to visit?")
```

```python
print(output_1)
```

```python
output_2 = conversation.predict(input="Which one has the best weather in winter?")
```

```python
print(output_2)
```
